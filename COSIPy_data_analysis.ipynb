```python
# Ensure the index is in datetime format
selected_df.index = pd.to_datetime(selected_df.index)

# Format the dates in the DataFrame
selected_df.index = selected_df.index.strftime('%Y-%m')

# Create a bar plot with the selected columns
ax = selected_df[df_bar].plot(kind='bar', stacked=True, zorder=2,legend=True)
ax.set_ylim([-450,450])
ax
# Define the markers and colors for the point plot
markers = ['2','1','^']  # Triangle up and triangle down
colors = ['black','black','pink']  # Red and blue
labels = ['MB', 'SMB', 'Î”SWE',"Precipitation (Khare)","Refreeze","Melt","Sublimation"]

# Add point data over it
for i, column in enumerate(df_point):
    ax.scatter(selected_df.index ,selected_df[column], marker=markers[i], color=colors[i], zorder=3,linewidth=0.8)
ax.legend(labels,frameon=False)  

# Define colors for each season
season_colors = {
    'monsoon': 'lightpink',
    'pre-monsoon': 'lightgreen',
    'post-monsoon': 'lightyellow',
    'winter': 'lightblue'
}

for i, month in enumerate(selected_df.index):
    season = selected_df.loc[month, 'season']
    if season in season_colors:
        rect = patches.Rectangle((i-0.5, ax.get_ylim()[0]), 1, ax.get_ylim()[1] - ax.get_ylim()[0], linewidth=0, facecolor=season_colors[season], alpha=0.3)
        ax.add_patch(rect)

ax.grid(True,alpha=0.2,linestyle='--')
#for i, month in enumerate(selected_df.index):
    #if selected_df.loc[month, 'season'] == 'monsoon':
        #rect = patches.Rectangle((i-0.5, ax.get_ylim()[0]), 1, ax.get_ylim()[1] - ax.get_ylim()[0], linewidth=0, facecolor='lightblue', alpha=0.3)
        #ax.add_patch(rect)

ax.grid(True,alpha=0.2,linestyle='--')
ax.set_ylabel("Mass fluxes [mm w.e.]", fontdict={'weight': 'bold'})
ax.set_xlabel("")
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig(r'C:\Users\Lenovo\Desktop\TC_brief\Figures\cosipy_output_low.png', dpi=200, facecolor='w', bbox_inches="tight")
plt.show()
```
